```python
###plot yaw-rot over time
yaw_rot = np.array(bot_stateDF["yaw_rot"])
timestep = np.array(bot_stateDF["timestep"])
print("length of yaw-rot: {}".format(len(yaw_rot)))
print("length of timestep: {}".format(len(timestep)))

plt.plot(timestep, yaw_rot)
plt.title("yaw_rot vs. timestep")
plt.xlabel("timeStep (s)")
plt.ylabel("yaw_rot")
plt.grid()
plt.show()
```

```python
print(np.mean(yaw_rot))

overshoot_arr = []
timestep_arr = []
for i in range(len(timestep)):
    if yaw_rot[i] > 0.01 or yaw_rot[i] < -0.01:
        timestep_arr.append(timestep[i])
        while(yaw_rot[i] > 0.01 or yaw_rot[i] < -0.01):
            overshoot_arr.append(yaw_rot[i])
            i += 1
            if i == len(timestep):
                break
    else:
        continue
if overshoot_arr:
    print(np.mean(overshoot_arr))
    print(timestep_arr[len(timestep_arr) - 1] - timestep_arr[0])
else:
    print("no overshoot found")
    
#if no overshoot found
##then increase gain until overshoot can be collected
```

```python
overshoot_exp = 0.002
time_exp = 0.3
time_act = timestep_arr[len(timestep_arr) - 1] - timestep_arr[0]
print(np.log((abs(np.mean(overshoot_arr))/overshoot_exp) + 1))
print(np.log((time_act/time_exp) + 1))

fitness = np.log((abs(np.mean(overshoot_arr))/overshoot_exp) + 1) + np.log((time_act/time_exp) + 1)
print(fitness)
```
